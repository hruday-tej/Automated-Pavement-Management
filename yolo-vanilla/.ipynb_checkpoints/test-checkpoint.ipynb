```python
class_names = ['pothole', 'alligator cracking', 'lateral cracking', 'longitudinal cracking']
```

```python
def plot_bounding_boxes_with_predictions(image_path, gt_labels_path, model):
    # Load image
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    height, width, _ = image.shape
    
    # Create two copies of the image for ground truth and prediction
    gt_image = image.copy()
    predicted_image = image.copy()
    
    # Read ground truth label file
    with open(gt_labels_path, 'r') as f:
        gt_labels = f.readlines()

    # Function to draw bounding boxes and class labels on an image
    def draw_bounding_boxes(image, labels, color, is_prediction=False):
        for label in labels:
            if not is_prediction:  # Ground truth
                class_id, x_center, y_center, bbox_width, bbox_height = map(float, label.strip().split())
                class_id = int(class_id)
            else:  # Prediction
                x_center, y_center, bbox_width, bbox_height, class_id = label[:5]
                class_id = int(class_id.cpu().numpy())  # Convert to integer class ID
            
            # Convert YOLO format (normalized) to pixel values
            x_center *= width
            y_center *= height
            bbox_width *= width
            bbox_height *= height
            
            # Calculate the top-left corner of the bounding box
            x_min = int(x_center - bbox_width / 2)
            y_min = int(y_center - bbox_height / 2)
            x_max = int(x_center + bbox_width / 2)
            y_max = int(y_center + bbox_height / 2)
            
            # Draw rectangle on the image
            cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)

            # Draw the class label text
            label_text = class_names[class_id]
            cv2.putText(image, label_text, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
        
        return image

    # Draw ground truth bounding boxes (blue) on the left image
    gt_image = draw_bounding_boxes(gt_image, gt_labels, (0, 0, 255))

    # Run model prediction
    results = model(image_path)  # Get the predictions from the model

    # Extract the bounding boxes from predictions
    predictions = results[0].boxes.xywhn  # YOLO outputs normalized (x_center, y_center, width, height)
    pred_classes = results[0].boxes.cls   # Get class IDs for predictions

    # Combine predictions with class IDs
    pred_data = []
    for i, pred in enumerate(predictions):
        pred_data.append(pred.tolist() + [pred_classes[i]])  # Add class ID to bbox data

    # Draw predicted bounding boxes (red) on the right image
    predicted_image = draw_bounding_boxes(predicted_image, pred_data, (255, 0, 0), is_prediction=True)

    # Plot the two images side by side
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # Display ground truth
    ax[0].imshow(gt_image)
    ax[0].set_title('Ground Truth')
    ax[0].axis('off')

    # Display predictions
    ax[1].imshow(predicted_image)
    ax[1].set_title('Predictions')
    ax[1].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
def get_random_image_and_label(image_dir, label_dir):
    image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpeg') or f.endswith('.jpg') or f.endswith('.png')]

    if not image_files:
        print("No image files found in the directory.")
        return None, None

    # Randomly select an image file
    random_image_file = random.choice(image_files)

    # Construct the corresponding label file path by replacing the image extension with .txt
    base_filename = os.path.splitext(random_image_file)[0]
    label_file = base_filename + '.txt'

    # Check if the corresponding label file exists in the label directory
    label_path = os.path.join(label_dir, label_file)
    if not os.path.exists(label_path):
        print(f"Label file not found for image {random_image_file}")
        return None, None

    # Construct the full path for the image and label
    image_path = os.path.join(image_dir, random_image_file)
    return image_path, label_path

image_dir = '/blue/cap4773/tej/Automated-Pavement-Management/data/processed/test/images/'
label_dir = '/blue/cap4773/tej/Automated-Pavement-Management/data/processed/test/labels/'

# Get a random image and its corresponding label
random_image_path, random_label_path = get_random_image_and_label(image_dir, label_dir)

plot_bounding_boxes_with_predictions(random_image_path, random_label_path, model)
```

```python
import torch
import numpy as np
```

```python
def calculate_iou(box1, box2):
    """
    Calculate Intersection over Union (IoU) between two bounding boxes.
    Each box is in the format [x_min, y_min, x_max, y_max].
    """
    x1_min, y1_min, x1_max, y1_max = box1
    x2_min, y2_min, x2_max, y2_max = box2

    # Calculate intersection coordinates
    x_inter_min = max(x1_min, x2_min)
    y_inter_min = max(y1_min, y2_min)
    x_inter_max = min(x1_max, x2_max)
    y_inter_max = min(y1_max, y2_max)

    # Calculate intersection area
    inter_area = max(0, x_inter_max - x_inter_min) * max(0, y_inter_max - y_inter_min)

    # Calculate areas of both boxes
    box1_area = (x1_max - x1_min) * (y1_max - y1_min)
    box2_area = (x2_max - x2_min) * (y2_max - y2_min)

    # Calculate union area
    union_area = box1_area + box2_area - inter_area

    # Return IoU
    return inter_area / union_area if union_area != 0 else 0
```

```python
def calculate_ap(precisions, recalls):
    """
    Calculate Average Precision (AP) given precision and recall values.
    """
    # Append zero to the precisions and recalls lists
    precisions = np.concatenate(([0.], precisions, [0.]))
    recalls = np.concatenate(([0.], recalls, [1.]))

    # Make precision values non-decreasing from right to left
    for i in range(len(precisions) - 1, 0, -1):
        precisions[i - 1] = np.maximum(precisions[i - 1], precisions[i])

    # Calculate AP by integrating area under curve (AUC)
    indices = np.where(recalls[1:] != recalls[:-1])[0]
    ap = np.sum((recalls[indices + 1] - recalls[indices]) * precisions[indices + 1])
    
    return ap
```

```python
def calculate_map(predictions, ground_truths, iou_threshold=0.5):
    """
    Calculate mean Average Precision (mAP) for a set of predictions and ground truths.
    
    Parameters:
    - predictions: List of predicted bounding boxes in the form (image_id, confidence, [x_min, y_min, x_max, y_max]).
    - ground_truths: List of ground truth bounding boxes in the form (image_id, [x_min, y_min, x_max, y_max]).
    - iou_threshold: IoU threshold to consider a prediction a true positive.

    Returns:
    - mAP: mean Average Precision
    """
    predictions = sorted(predictions, key=lambda x: x[1], reverse=True)  # Sort by confidence score
    tp = np.zeros(len(predictions))
    fp = np.zeros(len(predictions))

    gt_used = {img_id: np.zeros(len([gt for gt in ground_truths if gt[0] == img_id]))
               for img_id in set(gt[0] for gt in ground_truths)}

    for i, pred in enumerate(predictions):
        img_id, confidence, pred_box = pred[0], pred[1], pred[2:]
        gt_boxes = [gt[1:] for gt in ground_truths if gt[0] == img_id]

        ious = np.array([calculate_iou(pred_box, gt_box) for gt_box in gt_boxes])
        if len(ious) == 0 or np.max(ious) < iou_threshold:
            fp[i] = 1
        else:
            max_iou_idx = np.argmax(ious)
            if gt_used[img_id][max_iou_idx] == 0:
                tp[i] = 1
                gt_used[img_id][max_iou_idx] = 1
            else:
                fp[i] = 1

    tp_cumsum = np.cumsum(tp)
    fp_cumsum = np.cumsum(fp)

    precisions = tp_cumsum / (tp_cumsum + fp_cumsum)
    recalls = tp_cumsum / len(ground_truths)

    ap = calculate_ap(precisions, recalls)
    return ap
```

```python
def evaluate_model(model, test_images, test_labels, iou_threshold=0.5):
    """
    Function to evaluate the model and calculate both IoU and mAP.
    
    Parameters:
    - model: The trained YOLO model.
    - test_images: Path to test images.
    - test_labels: List of ground truth labels.
    - iou_threshold: IoU threshold to determine correct predictions.

    Returns:
    - mean_iou: Average IoU over all predictions.
    - mAP: mean Average Precision over all predictions.
    """
    # Predict with the model
    results = model(test_images)  # Model predicts on the test set
    predictions = results.xyxy  # Predicted boxes in format [x_min, y_min, x_max, y_max, confidence, class]

    # Prepare lists of predictions and ground truths
    preds = []
    for img_idx, pred in enumerate(predictions):
        for p in pred:
            preds.append((img_idx, p[4].item(), [p[0].item(), p[1].item(), p[2].item(), p[3].item()]))

    gts = []
    for img_idx, label in enumerate(test_labels):
        for gt in label:
            gts.append((img_idx, [gt[0], gt[1], gt[2], gt[3]]))  # Ground truths in [x_min, y_min, x_max, y_max] format

    # Calculate mean IoU over all images
    ious = []
    for pred in preds:
        pred_box = pred[2]
        gt_boxes = [gt[1] for gt in gts if gt[0] == pred[0]]
        if gt_boxes:
            iou_values = [calculate_iou(pred_box, gt_box) for gt_box in gt_boxes]
            ious.append(max(iou_values))  # Get the highest IoU per prediction

    mean_iou = np.mean(ious) if ious else 0

    # Calculate mAP
    mAP = calculate_map(preds, gts, iou_threshold)

    return mean_iou, mAP
```

```python
import os

def load_ground_truths(test_labels_path):
    """
    Load ground truth labels from the /test/labels directory.
    
    Each label file should be in YOLO format: [class_id, x_center, y_center, width, height].
    
    Parameters:
    - test_labels_path: Path to the directory containing the label files.
    
    Returns:
    - ground_truths: A list of ground truth boxes for each image.
    """
    ground_truths = []
    
    # Iterate over each label file
    for label_file in sorted(os.listdir(test_labels_path)):
        if label_file.endswith('.txt'):
            img_id = os.path.splitext(label_file)[0]  # Remove file extension
            with open(os.path.join(test_labels_path, label_file), 'r') as f:
                gt_boxes = []
                for line in f.readlines():
                    class_id, x_center, y_center, width, height = map(float, line.strip().split())
                    # Convert from YOLO format (x_center, y_center, width, height) to (x_min, y_min, x_max, y_max)
                    x_min = x_center - (width / 2)
                    y_min = y_center - (height / 2)
                    x_max = x_center + (width / 2)
                    y_max = y_center + (height / 2)
                    gt_boxes.append([x_min, y_min, x_max, y_max])
                ground_truths.append(gt_boxes)
    return ground_truths
```

```python
from ultralytics import YOLO

model = YOLO('./runs/detect/train15/weights/best.pt')
```

```python
test_images_path = '/blue/cap4773/tej/Automated-Pavement-Management/data/processed/test/images'  # Path to test images
test_labels_path = '/blue/cap4773/tej/Automated-Pavement-Management/data/processed/test/labels'

# Load the test images (list of file paths)
test_images = sorted([os.path.join(test_images_path, img) for img in os.listdir(test_images_path) if img.endswith(('.jpeg', '.png'))])
```

```python
# Load the ground truth labels from the label files
test_labels = load_ground_truths(test_labels_path)
```

```python
# Evaluate the model and calculate IoU and mAP
mean_iou, mAP = evaluate_model(model, test_images, test_labels)
```

```python

# Print the results
print(f"Mean IoU: {mean_iou:.3f}")
print(f"mAP@0.5: {mAP:.3f}")
```
